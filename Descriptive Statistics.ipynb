## T-Test

- 두 집단의 평균을 비교하여, 평균의 차이가 통게적으로 의미가 있는지를 검정

```python
import pandas as pd
import numpy as np

data1 = pd.read_csv('/JupyterNotbook/DataMining/winequality-red.csv', sep=';')
data2 = pd.read_csv('/JupyterNotbook/DataMining/winequality-white.csv', sep=';')
```

```python
print(data1.shape)
data1.head()
```

```python
print(data2.shape)
data2.head()
```

```python
data1.describe()
```

```python
data2.describe()
```

```python
from scipy import stats

stats.ttest_ind(data1['quality'], data2['quality'], equal_var=False)
```

## 분할표(Contingency Table)

- pandas의 crosstab 사용

```python
import seaborn as sns
sns.set_style('white')

titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.shape
```

```python
pd.crosstab(titanic['survived'], titanic['who'])
```

- margins=True: 총 합계

```python
pd.crosstab(titanic['survived'], titanic['who'], margins=True)
```

- normalize=True: 비율 값

```python
pd.crosstab(titanic['class'], titanic['who'], normalize=True)
```

- values: index와 column을 기준으로 한 values의 값을 표시
- aggfunc: 해당 함수로 계산 한 값 표시

```python
pd.crosstab(titanic['survived'], titanic['who'], values=titanic.fare, aggfunc=np.mean)
```

- 복수 컬럼

```python
pd.crosstab(titanic['survived'], [titanic['pclass'], titanic['who']], margins=True)
```

## 상관분석

### 결측치 처리

- 결측치 확인

```python
titanic.isnull().sum() # null값은 1, null이 아닌 값은 0 >> 유의미한 값 count
```

```python
print(titanic['embarked'].value_counts())
print(titanic['deck'].value_counts())
print(titanic['embark_town'].value_counts())
```

- 결측치 처리: .fillna()

```python
titanic.age = titanic.age.fillna(titanic.age.median())
titanic.embarked = titanic.embarked.fillna('S')
titanic.deck = titanic.deck.fillna('C')
titanic.embark_town = titanic.embark_town.fillna('Southampton')

titanic.isnull().sum()
```

### 피어슨 상관계수

```python
titanic_corr = titanic.corr(method='pearson')
titanic_corr
```

```python
titanic.survived.corr(titanic.adult_male) # method='pearson'
```

- heatmap()

```python
sns.heatmap(data=titanic_corr,
            annot=True, fmt='.2f', cmap='Blues')
```
