종료합니다.  


```python
park = []
out = []
n_cars = 0
space = 5

def pull_car(pull, push):
    last = pull.pop()
    push.append(last)
    

while True:
    inp = input('다음을 선택하세요. (0.종료 1.주차 2.출차 3.주차공간확인) ?')
    if inp == '0':
        print('종료합니다.')
        break
    elif inp == '1':
        if n_cars >= 5:
            print('주차공간이 없습니다.')
            continue
        else:
            new_car = input('차량번호를 입력하세요 ?')
            park.append(new_car)
            n_cars += 1
            space = 5 - n_cars
            print('차량번호', new_car, '주차 완료!', '주차대수:', n_cars)
    elif inp == '2':
        car = input('차량번호를 입력하세요. ?')
        if car in park:
            print('차량번호', car, '출차합니다.')
            n_out=n_cars-park.index(car)-1
            for i in range(n_out):
                pull_car(park, out)
            park.pop()
            for i in range(n_out):
                pull_car(out, park)
            n_cars = n_cars - 1
            space += 1
            print(n_out, '대 차량 재주차 완료!')
        else:
            print('차량번호', car, '은 주차 차량이 아닙니다.')
            continue
    elif inp == '3':
        print('주차가능 공간:', space)
        print('주차차량 대수:', n_cars)
        print('주차차량:')
        for  c in park:
            print(c)
    else:
        print('입력값은 0, 1, 2, 3 만 가능합니다.')
```
